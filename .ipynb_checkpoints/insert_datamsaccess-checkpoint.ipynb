```python
import pyodbc
import pandas as pd
```

```python
SHS = pd.read_csv('SHS.csv', encoding='utf-8').dropna(how='all')
SHS2 = pd.read_csv('SHS-2.csv', encoding='utf-8').dropna(how='all')
PRE = pd.read_csv('PRESCHOOL.csv', encoding='utf-8').dropna(how='all')
JHS = pd.read_csv('JHS.csv', encoding='utf-8').dropna(how='all')
# BSCRIM = pd.read_csv('SJTIT_BSCRIM.csv').dropna(how='all')
csvData = pd.concat([JHS, PRE, SHS2, SHS])
```

```python
with open('SHS.csv', 'rb') as f:
    content = f.read()

# Replace problematic bytes (e.g., byte `0xD1`) with a blank space or a compatible character
content = content.replace(b'\xD1', b' ')

# Write back to a temporary file or read it directly from memory
with open('SHS.csv', 'wb') as f:
    f.write(content)

```

```python
conn_str = (
    r'DRIVER={Microsoft Access Driver (*.mdb, *.accdb)};'
    r'DBQ=C:\Users\ISOTECH-DEV\Documents\Python\python-stc\HXData.mdb;'
)
conn = pyodbc.connect(conn_str)
crsr = conn.cursor()
```

```python
for index, row in csvData.iterrows():
    r = row
    id = r['IDNUMBER']
    rfid = int(r['RFID'])
    print(f'ID:{id} \t RFID:{rfid}')
```

```python

for index, row in csvData.iterrows():
    r = row
    id = r['IDNUMBER']
    rfid = r['RFID']
    print(id)
    sql = f'''INSERT INTO TEmployee (EmployeeCode, CardNo)
         VALUES ('{id}', '{rfid}')'''
    crsr.execute(sql)
conn.commit()


```
